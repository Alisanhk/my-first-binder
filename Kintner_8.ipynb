```python
#Import pandas
import pandas as pd
```

```python
#Import seaborn
import seaborn as sns
```

```python
#Read the avocado csv and display first 5 rows
avocado = pd.read_csv('avocado.csv')
avocado.head()
```

```python
#Display information about the dataframe
avocado.info()
```

```python
#Convert the Date column to datetime type
avocado['Date'] = pd.to_datetime(avocado['Date'])
```

```python
#Display dataframe information
avocado.info()
```

```python
avocado_1 = avocado.rename(columns={'type' : 'bag_type'})
```

```python
#Filter data to 2015, Conventional Type, Albany Region. 
#Store filtered data, only columns Date, Total Bags, and Small Bags in a new data frame. 
#Display first 5 rows
avocado_2015 = avocado_1.query('bag_type == "conventional"') \
    .query('region == "Albany"') \
    .query('Date >= "2015-01-01"') \
    .query('Date <= "2015-12-31"')
avocado_2015 = avocado_2015[['Date','Total Bags', 'Small Bags']]
avocado_2015.head()
```

```python
#Melt the Total Bags and Small Bags columns
#Name the column containing type of bags to Bags
#Name the column containing the counts to Count
#Display the new data frame
avocado_melt = avocado_1.melt(
    id_vars = 'Date',
    value_vars = ['Total Bags','Small Bags'],
    var_name = 'Bags',
    value_name = 'Count')
display(avocado_melt)
```

```python
#Plot the data into a line plot
sns.relplot(data=avocado_melt, kind ='line',x='Date',y='Count',hue='Bags')
```

```python
#Bin the total volume column into four quantiles and store the bin labels in a new column. 
#Display the first five rows. 
avocado_1 = avocado.rename(columns={'Total Volume' : 'Total_Volume'})
avocado_1['Sales Volume'] = pd.qcut(avocado_1.Total_Volume, q=4, 
                                    labels=['Poor','Modest','Good','Excellent'])
avocado_1.head()
```

```python
#Display data in count plot
sns.countplot(data=avocado_1, x="year", hue="Sales Volume")
```

```python
#Melt the three bag types
avocado_meltbags = avocado_1.melt(
    id_vars = 'year',
    value_vars = ['Small Bags','Large Bags','XLarge Bags'],
    var_name = 'Bags',
    value_name = 'Count')
display(avocado_meltbags)
```

```python
#Display number of bags for each bag type by year
sns.relplot(data=avocado_meltbags, kind ='line',x='year',y='Count',hue='Bags')
```
